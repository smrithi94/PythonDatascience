```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
```

```python
import numpy as np 
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,4,2)
plt.plot(y,x,'b')
plt.tight_layout()
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title('Bigger Plot')
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x,'r')
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='xpower2')
ax.plot(x,x**3,label='xpower3')
ax.legend(loc=0)
```
